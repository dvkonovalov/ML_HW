```python
# Подключение необходимых библотек и вспомогательные функцции
from scipy.linalg import solve
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    """
    Математическая функция из задания
    """
    return np.sin(x/5)*np.exp(x/10)+5*np.exp((-x)/2)
```

Сформирую систему линейных уравнений (то есть задам матрицу коэффициентов A и свободный вектор b) для многочлена первой степени, который должен совпадать с функцией f в точках 1 и 15. Решаю данную систему с помощью функции scipy.linalg.solve и нарисую функцию f и полученный многочлен

```python
A = np.array ([[1,1], [1,15]])
b = [f(1),f(15)]
z = solve(A, b)

def f1(x):
    """
    Функция приближения через многочлен первой степени
    """
    return z[0] + z[1]*x

x = np.arange(0, 16, 0.1)
plt.plot(x, f(x), color="r", label="f(x)")
plt.plot(x, f1(x), color="g", label="f1(x)")
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.legend()
plt.show()
```

ОТВЕТ НА ВОПРОС: Построенная функция плохо приближает заданную в задание функцию, так как является многочленом первой степени.

Повторю те же шаги для многочлена второй степени, который совпадает с функцией f в точках 1, 8 и 15.
?

```python
A = np.array ([[1,1, 1], [1,8, 64], [1, 15, 225]])
b = [f(1),f(8), f(15)]
z = solve(A, b)

def f2(x):
    """
    Функция приближения через многочлен второй степени
    """
    return z[0] + z[1]*x + z[2]*x**2

x = np.arange(0, 16, 0.1)
plt.plot(x, f(x), color="r", label="f(x)")
plt.plot(x, f2(x), color="g", label="f2(x)")
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.legend()
plt.show()
```

ОТВЕТ НА ВОПРОС: качество апроксимации улучшилось, но совсем незначительно.

Повторю те же шаги для многочлена третьей степени, который совпадает с функцией f в точках 1, 4, 10 и 15

```python
A = np.array ([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]])
b = [f(1),f(4), f(10), f(15)]
z = solve(A, b)

def f3(x):
    """
    Функция приближения через многочлен третьей степени
    """
    return z[0] + z[1]*x + z[2]*x**2 + z[3]*x**3

x = np.arange(0, 16, 0.1)
plt.plot(x, f(x), color="r", label="f(x)")
plt.plot(x, f3(x), color="g", label="f3(x)")
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.legend()
plt.show()
print("Ответ на задачу - ", *z)
```

ОТВЕТ НА ВОПРОС: многочлен третьей степени хорошо апроксимирует заданную условием функцию.

Апроксимация нужна если необходимо заменить сложную функцию более простой и легкой в вычислениях. Примеры задач, где нужна апроксимация - разработка компьютерных программ для сложных вычислений, исследование числовых характеристик и качественных свойств объекта и т.д.
Существует множество различных метрик определения эффективности апроксимации. Некоторые из них: среднеквадратичная ошибка, корень из среднеквадратичной ошибки, среднеквадратичная ошибка, средняя ошибка и другие.
