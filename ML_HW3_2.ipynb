Возьму датасет с сортами вина из sklearn

```python
from sklearn.datasets import load_wine
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np

data = load_wine()
X = pd.DataFrame(data=data['data'],columns=data['feature_names'])
Y = pd.DataFrame(data=data['target'], columns=['target'])
```

Использую метод KFold для деления выборки на тренировочную и тестовую. Для воспроизводимости зафиксирую параметр random_state=42 и shuffle=True, так как иначе будет ошибка.
Для KFold и для каждого k ∈ [1, 50] 
(число "соседей") прогнюь алгоритм ближайших соседейиз (sklearn.neighbor 
и посчитать долю правильных ответоват?

```python
from sklearn.model_selection import KFold

kfold = KFold(random_state = 42, shuffle=True)
for train_indices, test_indices in kfold.split(X):
    X_train, X_test, y_train, y_test = X.iloc[train_indices], X.iloc[test_indices], Y.iloc[train_indices], Y.iloc[test_indices]
y_train = np.array(y_train['target'])
y_test = np.array(y_test['target'])
k_best = [0, 0]
for k in range(1, 51):
    classifier = KNeighborsClassifier(n_neighbors = k)
    classifier.fit(X_train, y_train)
    y_predicted = classifier.predict(X_test)
    print(f'for k = {k} the accuracy -', round(np.mean(y_test==y_predicted), 3))
    if round(np.mean(y_test==y_predicted), 3) > k_best[1]:
        k_best = [k, round(np.mean(y_test==y_predicted), 3)]

print(f'Лучший результат равен {k_best[1]} при k = {k_best[0]}')
```

Результат для KFold достаточно неплох, равен 82,9% и достиг максимума при k=3.
Теперь проведем все тоже самое, но разделим выборку с помощью метода LOO

```python
from sklearn.model_selection import LeaveOneOut

loo = LeaveOneOut()
for train_indices, test_indices in loo.split(X):
    X_train, X_test, y_train, y_test = X.iloc[train_indices], X.iloc[test_indices], Y.iloc[train_indices], Y.iloc[test_indices]
y_train = np.array(y_train['target'])
y_test = np.array(y_test['target'])
k_best = [0, 0]
for k in range(1, 51):
    classifier = KNeighborsClassifier(n_neighbors = k)
    classifier.fit(X_train, y_train)
    y_predicted = classifier.predict(X_test)
    print(f'for k = {k} the accuracy -', round(np.mean(y_test==y_predicted), 3))
    if round(np.mean(y_test==y_predicted), 3) > k_best[1]:
        k_best = [k, round(np.mean(y_test==y_predicted), 3)]

print(f'Лучший результат равен {k_best[1]} при k = {k_best[0]}')
```

LOO работает уже более интересно. Мы получаем или 100% точность или нолевую. При этом алгоритм очень быстро достиг 100% точности(уже при k=1), что несомненно является его достоинством в данном случае.
Теперь прогоним тот же алгоритм, но с делением выборки с помощью метода StratifiedKFold

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(random_state=42, shuffle=True)
for train_indices, test_indices in skf.split(X, Y):
    X_train, X_test, y_train, y_test = X.iloc[train_indices], X.iloc[test_indices], Y.iloc[train_indices], Y.iloc[test_indices]
y_train = np.array(y_train['target'])
y_test = np.array(y_test['target'])
k_best = [0, 0]
for k in range(1, 51):
    classifier = KNeighborsClassifier(n_neighbors = k)
    classifier.fit(X_train, y_train)
    y_predicted = classifier.predict(X_test)
    print(f'for k = {k} the accuracy -', round(np.mean(y_test==y_predicted), 3))
    if round(np.mean(y_test==y_predicted), 3) > k_best[1]:
        k_best = [k, round(np.mean(y_test==y_predicted), 3)]

print(f'Лучший результат равен {k_best[1]} при k = {k_best[0]}')
```

StratifiedKFold работает хуже, чем kFold и LOO в данном случае, так как при нем достигается наименьшая точность (77%). Также можно отметить, что наилучшая точность достигается при k=6
Если обобщить результаты, то можно сказать, что  метод кросс-валидации LOO дает наилучший из трех вариантов результат в 100% за наименьшее значение k равное 1 (при этом при методе LOO значение k также минимальное).?

Произведу масштабирование признаков с помощью функции sklearn.preprocessing.scaleи?

```python
from sklearn.preprocessing import scale

data = load_wine()

X = pd.DataFrame(data=scale(data['data']),columns=data['feature_names'])
```

Снова найду оптимальное k на трех разных кросс-валидациях рассмотренных выше. Начну с KFold

```python
kfold = KFold(random_state = 42, shuffle=True)
for train_indices, test_indices in kfold.split(X):
    X_train, X_test, y_train, y_test = X.iloc[train_indices], X.iloc[test_indices], Y.iloc[train_indices], Y.iloc[test_indices]
y_train = np.array(y_train['target'])
y_test = np.array(y_test['target'])
k_best = [0, 0]
for k in range(1, 51):
    classifier = KNeighborsClassifier(n_neighbors = k)
    classifier.fit(X_train, y_train)
    y_predicted = classifier.predict(X_test)
    print(f'for k = {k} the accuracy -', round(np.mean(y_test==y_predicted), 3))
    if round(np.mean(y_test==y_predicted), 3) > k_best[1]:
        k_best = [k, round(np.mean(y_test==y_predicted), 3)]

print(f'Лучший результат равен {k_best[1]} при k = {k_best[0]}')
```

В случае использования KFold теперь мы смогли достичь 100% точность при k=9 (для сравнения до этого было 83% и k=3), то есть результат несомненно улучшился хоть и увеличился k.
Теперь проганю алгоритм на LOO

```python
from sklearn.model_selection import LeaveOneOut

loo = LeaveOneOut()
for train_indices, test_indices in loo.split(X):
    X_train, X_test, y_train, y_test = X.iloc[train_indices], X.iloc[test_indices], Y.iloc[train_indices], Y.iloc[test_indices]
y_train = np.array(y_train['target'])
y_test = np.array(y_test['target'])
k_best = [0, 0]
for k in range(1, 51):
    classifier = KNeighborsClassifier(n_neighbors = k)
    classifier.fit(X_train, y_train)
    y_predicted = classifier.predict(X_test)
    print(f'for k = {k} the accuracy -', round(np.mean(y_test==y_predicted), 3))
    if round(np.mean(y_test==y_predicted), 3) > k_best[1]:
        k_best = [k, round(np.mean(y_test==y_predicted), 3)]

print(f'Лучший результат равен {k_best[1]} при k = {k_best[0]}')
```

В случае LOO и масштабирования признаков сложно судить об улучшение/ухудшение алгоритма, так как точность и значение k в прошлом случае были "максимальными" из возможных и увидеть разницу не представляется возможным.
Теперь проганю алгоритм на методе StratifiedKFold

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(random_state=42, shuffle=True)
for train_indices, test_indices in skf.split(X, Y):
    X_train, X_test, y_train, y_test = X.iloc[train_indices], X.iloc[test_indices], Y.iloc[train_indices], Y.iloc[test_indices]
y_train = np.array(y_train['target'])
y_test = np.array(y_test['target'])
k_best = [0, 0]
for k in range(1, 51):
    classifier = KNeighborsClassifier(n_neighbors = k)
    classifier.fit(X_train, y_train)
    y_predicted = classifier.predict(X_test)
    print(f'for k = {k} the accuracy -', round(np.mean(y_test==y_predicted), 3))
    if round(np.mean(y_test==y_predicted), 3) > k_best[1]:
        k_best = [k, round(np.mean(y_test==y_predicted), 3)]

print(f'Лучший результат равен {k_best[1]} при k = {k_best[0]}')
```

В случае StratifiedKFold и масштабирования признаков результат несомненно улучшился. Теперь алгоритм достигает 100% точности (для сравнения до этого было 77%) при значение k=4 (до этого k=6, то есть и здесь есть улучшения)

Если подвести итог результатов после масштабирования признаков, то можно сказать, что масштабирование имеет положительный результат для улучшения точности и уменьшения k в некоторых методах (например - StratifiedKFold).
При этом оптимальный метод валидации не изменился в виду того, что и до этого он имел высокие показатели.
